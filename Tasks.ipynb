# Задание 6.

Пользователи из каких стран совершили наибольшее число успешных бронирований? Бронирование считается успешным, если в дальнейшем не было отменено (переменная is_canceled). В качестве ответа выберите страны, входящие в топ-5.

Подсказка: здесь необходимо сделать группировку по ['country'] и агрегирование по сумме 'is_canceled'.


```python
result = bookings.groupby('country').agg(total_canceled=('is_canceled','sum'))
result = result.sort_values(by='total_canceled',ascending=True)
next_result = result.sort_values(by='total_canceled',ascending=False)
print(result.head(5))
print(next_result.head(5))
```

# Задание 9.

Теперь проанализируйте даты запланированного прибытия (arrival_date_year). На какой месяц чаще всего оформляли бронь в 2016 году? Изменился ли самый популярный месяц в 2017?

Вам пригодится arrival_date_month.agg(pd.Series.mode) - Mode - возвращает самое часто встречаемое значение, если таких несколько возвращает их все


```python
bookings_2016 = bookings[bookings['arrival_date_year']==2016]
mounth_2016 = bookings_2016.groupby('arrival_date_month',as_index=False).size()
print('За 2016 год список популярных месяцев')
print(mounth_2016)
print()
bookings_2017 = bookings[bookings['arrival_date_year']==2017]
mounth_2017 = bookings_2017.groupby('arrival_date_month',as_index=False).size()
print('За 2017 год список популярныз месяцев')
print(mounth_2017)
```

# Задание 10.

Сгруппируйте данные по годам, а затем проверьте, на какой месяц (arrival_date_month) бронирования отеля типа 
City Hotel отменялись чаще всего в 2015? 2016? 2017?

Подсказка: для этого сначала примените метод groupby, после чего выберите нужную колонку для дальнейших действий

Например:

df.groupby('column_1')['column_2'].what_to_do()

```python
canceled_bookings = (
    bookings
    .groupby(['arrival_date_year', 'hotel'])  # Группируем по году и отелю
    ['is_canceled']  # Выбираем столбец с отменами
    .sum()           # Суммируем все значения 1 (каждая отмена = 1)
    .reset_index()   # Преобразуем индекс в обычные столбцы
    .rename(columns={'is_canceled': 'total_cancelations'})  # Переименовываем колонку
)

print(canceled_bookings)
```
